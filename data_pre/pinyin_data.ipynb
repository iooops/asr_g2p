# thchs30

```python
base_dir = '/home/xingxing/tts/dataset/data_thchs30/data'
thchs30list = os.listdir(base_dir)
```

```python
def readfile(path):
    fo = open(path, "r")
    lines = fo.readlines()
    fo.close()
    return lines
```

```python
thchs30_list = []
all_pinyins = set()

for t in thchs30list:
    # print(t)
    if t.endswith('.trn'):
        lines = readfile(base_dir + '/' + t)
        wav_file = base_dir + '/' + t[:-4]
        text = lines[0].strip().replace(' ', '')
        pinyins = lines[1].strip().split(' ')
        all_pinyins.update(pinyins)
        thchs30_list.append((wav_file, text, pinyins))
```

```python
sorted(list(all_pinyins))
```

```python
thchs30_list[:2]
```

```python
with open("thchs30_list.json", "w") as file:
    json.dump(thchs30_list, file, indent=4, sort_keys=True, ensure_ascii=False)
```

# BIAOBEI

```python
blist = readfile('/home/xingxing/tts/dataset/BZNSYP/ProsodyLabeling/000001-010000.txt')
```

```python
biaobei_list = []
for i in range(int(len(blist)/2)):
    filename, text = blist[2*i].strip().split('\t')
    pinyins = blist[2*i+1].strip().split(' ')   
    if not 'P' in pinyins:
        # print(filename, text, pinyins)
        all_pinyins.update(pinyins)
        filepath = '/home/xingxing/tts/dataset/BZNSYP/Wave/' + filename + '.wav'
        text = text.replace('#1', '').replace('#2', '').replace('#3', '').replace('#4', '')
        biaobei_list.append((filepath, text, pinyins))
```

```python
len(biaobei_list)
```

```python
with open("biaobei_list.json", "w") as file:
    json.dump(biaobei_list, file, indent=4, sort_keys=True, ensure_ascii=False)
```

# aishell3

```python
aishell3_basedir = '/home/xingxing/tts/dataset/aishell3'

as_train_list = readfile(os.path.join(aishell3_basedir, 'train/content.txt'))
as_test_list = readfile(os.path.join(aishell3_basedir, 'test/content.txt'))

as_lists = {
    'train': as_train_list,
    'test': as_test_list
}
```

```python
aishell3_list = []

for k in as_lists:    
    for al in as_lists[k]:
        filename, text = al.split(maxsplit=1)
        # print(filename, text)
        filepath = os.path.join(aishell3_basedir, k, 'wav', filename[:7], filename)
        ts = text.strip().split(' ')
        content = ''.join(ts[::2])
        pinyins = ts[1::2]
        all_pinyins.update(pinyins)
        # print(content, pinyins)
        aishell3_list.append((filepath, content, pinyins))
```

```python
with open("aishell3_list.json", "w") as file:
    json.dump(aishell3_list, file, indent=4, sort_keys=True, ensure_ascii=False)
```
